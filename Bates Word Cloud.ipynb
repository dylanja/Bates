```python
bates = "Who We Are Bates Group has been a trusted partner in the financial services industry for over 30 years. Our mission is to help our clients resolve and prevent financial and reputational risks. We do this by providing subject matter experts from our extensive nationwide roster of consultants. We place an intentional focus on recruiting, retaining and developing high-quality people and aspire to foster a culture of mutual respect and collaboration while being the preferred consulting firm for clients needing first-rate analytics and industry expert knowledge. Position Summary We are seeking an entry level Data Analyst to join our consulting firm. If you have a passion for math or an interest in financial analysis or the stock market this might be the opportunity for you! The successful Data Analyst has a proven aptitude for detail, an interest in learning new technologies and a desire to learn and grow within the consulting industry. In this support role, you will work on multiple projects analyzing client investment data, ensuring accuracy and looking for discrepancies on confidential financial litigation matters for the purpose of creating profit & loss analysis and various other investment related schedules. Key Responsibilities Utilize Bates proprietary program to construct investment analysis and understand the underlying function of basic investment related schedules. Data entry: inputs case data into standardized electronic format, preparing case matter schedules and analyses. Review data: ensures the accuracy of trading and cash activity; reviews for unusual transactions and discrepancies. Required Skills Intermediate Excel Skills Must be able to work overtime, evenings and weekends as projects demand Ability to prioritize and remain organized in a fast paced, deadline-oriented environment Accustomed to working independently and as part of a team; while interacting with all levels of staff and management Desired Skills Advanced Excel skills Financial acumen gained through prior experience in accounting, investment, or financial services BS or BA in quantitative field: mathematics, economics, finance, statistics, accounting"
```

```python
stopwords = set(STOPWORDS)
```

```python
bates_wc = WordCloud(
    background_color='lightgrey',
    max_words=2000,
    stopwords=stopwords)

bates_wc.generate(bates)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(bates_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
